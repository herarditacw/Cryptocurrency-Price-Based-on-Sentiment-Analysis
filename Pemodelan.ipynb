```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eth = pd.read_csv(r"C:\Users\herar\Downloads\Historical Price\eth-usdt.csv")
eth
```

```python
btc = pd.read_csv(r"C:\Users\herar\Downloads\Historical Price\btc-usdt.csv")
btc
```

```python
plt.plot(eth['Close Time'], eth['Close'])
plt.xlabel('Close Time')
plt.ylabel('Close Price')
plt.title('Ethereum Close Time and Close Price')
plt.show()
```

```python
plt.plot(btc['Close Time'], btc['Close'])
plt.xlabel('Close Time')
plt.ylabel('Close Price')
plt.title('Bitcoin Close Time and Close Price')
plt.show()
```

```python
sns.relplot(x='Close Time', y='Close', data = eth, kind='line')
plt.show()
```

```python
sns.relplot(x='Close Time', y='Close', data = btc, kind='line')
plt.show()
```
